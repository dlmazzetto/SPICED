```python
import matplotlib.pyplot as plt
%matplotlib inline 
import pandas as pd
```

```python
# Load data
penguins = pd.read_csv("penguins_simple.csv", sep= ";")
penguins.head()
```

# Plotting with Seaborn

```python
import seaborn as sns
 
sns.scatterplot(data = penguins, x ='Body Mass (g)', y = 'Flipper Length (mm)', alpha= 0.8)
sns.despine()
```

### Seaborn makes it easy to color by species

```python
sns.scatterplot(data = penguins, x = 'Culmen Length (mm)', y = 'Culmen Depth (mm)', hue = 'Species')
```

```python
sns.scatterplot(data = penguins, x = 'Culmen Length (mm)', y = 'Culmen Depth (mm)')
```

### Histogram 

```python
# In Seaborn 
sns.distplot(penguins['Body Mass (g)'])
```

```python
sns.pairplot(data = penguins, hue = 'Species')
```
