# Create DataFrames

```python
import pandas as pd
```

### Method 1: Nested List

```python
data = [['Logistic Lemongrass', 15],
        ['Jupyter Jasmine', 13],
        ['Allspice Arrays', 15]
       ]

df = pd.DataFrame(data, index=['A', 'B', 'C'], columns=['name', 'students'])
```

```python
df
```

### Method 2: Dictionary

```python
data = {
    'name': ['Logistic Lemongrass', 'Jupyter Jasmine', 'Allspice Arrays'],
    'students': [15, 13, 15]
}

df = pd.DataFrame(data, index=['A', 'B', 'C'])
df
```
