```python
x = np.linspace(-10, 10, num=41)
type(x)
```

#### Vectorization
**With Numpy we have to write fewer for loops**

```python
y = x ** 2
```

```python
# alternative:
# np.power(x, 2)
```

```python
len(x), len(y)
```

```python
x.shape, y.shape
```

```python
plt.figure(figsize=(10, 7)) # start new figure, figsize in inches

# plots with 1 Row, 2 columns, this is the 1st plot
plt.subplot(1, 2, 1)
plt.plot(x, (x**3 - x**2) * 0.1, color="forestgreen", label="polynomial") # 140 color names
plt.plot(x, np.sin(x) * 50 + 10, color="#c0ffee", label="sine")    # RGB hex codes
plt.legend(loc='lower center')  # <-- alternative: put labels here

# plots with 1 Row, 2 columns, this is the 2nd plot
plt.subplot(2, 2, 4)
plt.plot(x, y, '--', color=(1.0, 0, 0.75), label='parabola') # RGB tuple
# alternative: define a palette

plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial $y = x^2$')
plt.legend()  # <-- alternative: put labels here

plt.savefig('parabola.png', dpi=150)  # dots per inch: 1500 x 1050 pixels
```
