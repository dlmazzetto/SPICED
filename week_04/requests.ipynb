# requests

requests is one of the most loaded Python libraries. It allows you to send requests to websites.

```python
# pip install requests
```

```python
import requests
```

```python
# Choose the URL you want to request
url = 'https://www.lyrics.com/artist/Beyonc%C3%A9/349078'
```

```python
# Send the request
request = requests.get(url)
```

```python
# Check the status code in order to assess wether our request was succesful
request.status_code
```

200 = successful

300 = redirect

Everything with a 4 in the beginning will be an error on your side
400 = Error
401 = Not authorized
404 = webpage cannot be found

Everything with a 5 in the beginning will be an error on the server side
500

```python
# It is a requests type
type(request)
```

```python
# requests.text will return the html file of the website as a string
print(request.text)
```

```python
type(request.text)
```

```python
# Save the html in a txt file and search in an editor
with open('Beyonce.txt', 'w') as file:
    file.write(request.text)
```
