```python
dates = ['2020-01-01', '2020-01-04', '2020-01-05', '2020-01-09', '2020-01-11',
         '2020-01-01', '2020-01-07', '2020-01-11']
places = ['berlin', 'oslo', 'belgrade', 'berlin', 'checkout',
          'lisbon', 'beijing', 'checkout']
person = [1, 1, 1, 1, 1, 2, 2, 2]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'place': places, 'id': person}, index=pd.to_datetime(dates))
```

```python
df = df.groupby('id').resample('D').ffill()
```

```python
df.drop('id', axis=1, inplace=True)
```

```python
df = df.reset_index().set_index('level_1')
```

```python
df['before'] = df['place'].shift(1)
df
```

```python
df = df[df['before'] != 'checkout']  # could replace by 'entrance'
df.fillna('entrance', inplace=True)
```

```python
df.groupby(['place', 'before']).count()['id'].unstack()
```

```python
pd.crosstab(df['place'], df['before'], normalize=0) # <-- 0 axis
```
