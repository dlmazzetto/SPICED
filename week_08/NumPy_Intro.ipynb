## Creating NumPy Arrays

```python
data = []
for i in range(1, 10):
    if i % 2:
        d = i ** 2
        data.append(d)
data
```

```python
def square(x):
    return x ** 2
```

```python
data = [square(i) for i in range(1, 10) if i % 2]
data
```

## Two-Dimensional arrays

```python
a = np.zeros((100, 400), dtype=np.uint8) # <-- 0 .. 255
a
```

```python
from PIL import Image   # pip install pillow
```

```python
Image.fromarray(a)
```

```python
a.shape
```

```python
b = np.random.randint(0, 255, size=(100, 400), dtype=np.uint8)
Image.fromarray(b)
```

#### 6. change the shape

```python
c = b.reshape(200, 200)
c.shape
```

```python
Image.fromarray(c)
```

```python
np.arange(25)
```

```python
np.arange(25).reshape(5, 5)
```

```python
im = Image.open('tram.jpg')
d = np.array(im)
im
```

```python
d.shape
```

#### 7. Crop / slice the array

```python
e = d[25:, :-25, :]
e.shape
```

```python
Image.fromarray(e)
```

```python
# slice off the red layer
f = e[:, :, 0]
f.shape
```

```python
Image.fromarray(f)
```

```python
g = e.copy()
g[:, :, 1] = 0  # set green to zero
g[:, :, 2] = 0  # set blue to zero
Image.fromarray(g)
```

```python
h = f.reshape(100, 400)  # every 2nd row -> left, every 2nd to the right half
Image.fromarray(h)
```

```python
j = e.copy()
```

```python
j[20:80, :, 2] = 255
j[:, 50:150, 1] = 0
```

```python
Image.fromarray(j)
```

```python
k = e.copy()
k[:100, :100, :] = k[100:, 100:, :]
Image.fromarray(k)
```
